```python
import text_metrics.base
from text_metrics.resource_pool import rp as default_rp

# 1. El texto que quieres analizar
texto_pix = "fiz um pix mas o meu dinheiro nao caiu na conta"

# 2. Creamos el objeto de texto que NILC-Metrix entiende
t = text_metrics.base.Text(texto_pix)

# 3. Ejecutamos el motor de métricas
# Esto usará tu base de datos local y NLPNet automáticamente
print("Analizando texto... por favor espera.")
try:
    results = text_metrics.CMP_METRICS.values_for_text(t, default_rp).as_flat_dict()

    # 4. Imprimimos los resultados de forma ordenada
    print("\n--- RESULTADOS DEL ANÁLISIS ---")
    for metric, value in results.items():
        # Solo imprimimos métricas con valor para no saturar la pantalla
        if value != 0 and value is not None:
            print(f"{metric}: {value}")

except Exception as e:
    print(f"Error durante el análisis: {e}")
```
